# EXP_8 - Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:\\Users\\RAHUL\\OneDrive\\Desktop\\IDS_PRACTICAL\\healthcare_test_data.csv - healthcare_test_data.csv.csv")
df.head()
```

# 1. Scatterplot

```python
plt.scatter(df['Available Extra Rooms in Hospital'],df['Visitors with Patient'])
plt.xlabel('Available Extra Rooms in Hospital')
plt.ylabel('Visitors with Patient')
plt.title('Relation between Available Extra Rooms in Hospital and Visitors with Patient')
plt.show()
```

```python
sns.displot(df['City_Code_Hospital'],bins = 20 ,kde = False)
plt.title('Histogram')
plt.xlabel('City_Code_Hospital')
plt.ylabel('Frequency')
plt.show()
```

# BarCharts

```python
sns.barplot(x='Hospital_type_code' , y = 'City_Code_Hospital' , data = df)
```

# PieChart

```python
severty_counts = df['Severity of Illness'].value_counts()
plt.pie(severty_counts.values , labels = severty_counts.index , autopct = '%1.1f%%')
plt.title('Distribution of Severty of Illness')
plt.show()
```

# Linegraphs


```python
sns.lineplot(x = 'Hospital_type_code' , y = 'City_Code_Hospital' , data = df)
```
